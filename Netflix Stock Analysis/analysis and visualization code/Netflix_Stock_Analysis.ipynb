```python
#import the necessary libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
#load the csv file
data = pd.read_csv("NFLX.csv")
```

```python
#see the first 10 data
data.head(10)
```

```python
#set the default size of the figure width = 10 inch, height = 5 inch
sns.set(rc = {'figure.figsize':(10,5)})
```

```python
#convert the date dataframe from string to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
#Date column set as index column
data=data.set_index('Date')
```

```python
data.head(10)
```

#Objective 1 : Volume of Stock Traded

```python
#here I want to plot the graph between volume and year
#I use the lineplot to visualize this

sns.lineplot( x=data.index, y=data.Volume, color='red' )
plt.title('Volume of Stock Traded')

#to save the figure
plt.savefig('Volume_of_stock_traded.png')

plt.show()
```

#Objective 2 : Netflix Stock price high, open, low, close 

```python
data.head(10)
```

```python
data.plot( y=['High', 'Close', 'Open', 'Low'])  
#it automatically uses index as x axis value

plt.title('Netflix Stock price high, open, low, close')

#to save the figure
plt.savefig('Netflix_Stock_price_holc.png')

plt.show()
```

#Objective 3 : Netflix Stock price Day, Month, year wise

```python
data.head(10)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize = (15,10))
#This line creates a 3x1 subplot grid (plt.subplots(3))
#unpacks the three individual subplot axes (ax1, ax2, ax3) from the fig object. 

#here its grouping the day component of the date time index and plot the mean values of a day   
data.groupby(data.index.day).mean().plot(y='Volume', ax=ax1, xlabel='Day', color='green')

#here its grouping the day component of the date time index and plot the mean values of a month
data.groupby(data.index.month).mean().plot(y='Volume', ax=ax2, xlabel='Month', color='blue')

#here its grouping the day component of the date time index and plot the mean values of a year
data.groupby(data.index.year).mean().plot(y='Volume', ax=ax3, xlabel='year', color='orange')

# Adjust the vertical space between subplots
plt.subplots_adjust(hspace=1)  

#called to automatically adjust subplot parameters to prevent overlapping.
plt.tight_layout()

#to save the figure
plt.savefig('Netflix_Stock_price_dmy.png')

plt.show()
```

#Objective 4 & 5 : Top 5 Dates with highest stock price & top 5 dates with lowest stock price

```python
#top 5 dates of highest stock price
#sort the high column in decending order
a = data.sort_values(by='High',ascending=False)
a.head()
```

```python
#top 5 data of lowest stock price
#sort the high column in decending order
a = data.sort_values(by='Low',ascending=True)
a.head()
```

```python
#create the two plot(column) on same x axis
fig, axes = plt.subplots(nrows=1,ncols=2,sharex=True,figsize=(12,5))

#set the title 
fig.suptitle('High & low value stock per period of time',fontsize=18)

#first plot the higher stock value
sns.lineplot(ax=axes[0],y=data['High'],x=data.index,color='green')

#it is the plot of lower stock values
sns.lineplot(ax=axes[1],y=data['Low'],x=data.index,color='red')

#to save the figure
plt.savefig('top_5_dates_hl.png')

plt.show()
```
